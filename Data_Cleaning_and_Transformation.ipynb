```python
#Import relevant Py libraries to clean and transform data
import pandas as pd
import numpy as np
```

```python
#To read the csv file and have a glimpse of the dataset
df = pd.read_csv('imdb_top_1000.csv')
df.head()
```

```python
#To have the basic info about the dataset(the data has 16 columns and 1000 rows of different set information)
df.info()
```

```python
#To locate columns with missing values (data shows that some important info like gross, meta score is missing.)(We need to find solution for these missing values)
df.isnull().sum()
```

```python
#To count mean value of gross (income). It is important to find mean because we can use this to fill the missings values in column, Gross.
# Convert 'Gross' column to numeric, handling errors
df['Gross'] = pd.to_numeric(df['Gross'].str.replace(',', '').str.replace('$', ''), errors='coerce')
df.Gross.mean()
```

```python
#To convert NaN values into 0 (it will later on help converting float into integer)
df['Gross'] = df['Gross'].fillna(0)

#To convert mean float into integer (it is important to do this to better merge with the values in the given column)
df['Gross'] = df['Gross'].astype(int)
df.head(10)
```

```python
#Replace gross 0 (missing value) with the mean value
df['Gross'] = df['Gross'].replace(0, 68034750)
df.head(10)
```

```python
#Movie with maximum gross earnings
df.loc[df['Gross'].idxmax()]
```

```python
#Movie with minimum gross earnings
df.loc[df['Gross'].idxmin()]
```

```python
#Filling missing values in column "Meta_score"
df['Meta_score'] = df['Meta_score'].fillna(0)
df.head(10)
```

```python
#Filling missing values (0) in "Meta_score" with mean of the total
#First lets find the mean value
df['Meta_score'].mean()
```

```python
#To fill the missing value with mean
df['Meta_score'] = df['Meta_score'].replace(0, 66)
df.head(10)
```

```python
# To calculate correlation among different variables
numeric_df = df.select_dtypes(include=['number'])
correlation = numeric_df.corr()
print(correlation)
```

```python
#Finding top ten movies in terms of "Gross"
df.sort_values(by='Gross', ascending=False).head(10)[['Series_Title', 'Gross']]
```

```python
#Finding top ten movies in terms of "Rating"
df.sort_values(by='IMDB_Rating', ascending=False).head(10)[['Series_Title', 'IMDB_Rating']]
```
